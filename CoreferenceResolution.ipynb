## Run

Import "articles_dict.pickle" before running the following cells

```python
from tqdm.auto import tqdm
import pickle
```

```python
import spacy
import neuralcoref

coref = spacy.load('en_core_web_sm') #ner = spacy.load('en_core_web_lg')
neuralcoref.add_to_pipe(coref)
```

```python
with open("/content/articles_dict.pickle", "rb") as file:
    articles_dict = pickle.load(file)
```

```python
articlesDict_new = dict()
```

```python
for key in tqdm(articles_dict):
    if articles_dict[key]:
        doc = coref(articles_dict[key])
        articlesDict_new[key] = doc._.coref_resolved
    else:
        articlesDict_new[key] = None
```

```python
with open("new_articles_dict.pickle", "wb") as file:
    pickle.dump(articlesDict_new, file)
```
